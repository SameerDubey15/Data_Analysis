## Handling the null values in the Height and Weight column

```python
ath.Height.describe()
```

```python
ath.Weight.describe()
```

```python
# Fill all the NaN values with the mean value of the weight and height column
ath["Height"].fillna(ath.Height.mean(), inplace = True)
ath["Weight"].fillna(ath.Weight.mean(), inplace = True)
```

```python
# Correcting Datatype
ath.Height = ath.Height.astype('float')
ath.Weight = ath.Weight.astype('float')
```

```python
ath.info()
```

## Remove the unwanted phrases from the "Event" Column

```python
ath[['Sport', 'Event']]
```

We can see that sports column contains the name of sports and event column also contains the name of sports but with some event name. So we have to remove sportname from event column

```python
for i,j in zip(ath.Sport, range(len(ath.Event))):
  ath.Event[j] = re.sub(f'^{i}\s', '', ath.Event[j])
```
